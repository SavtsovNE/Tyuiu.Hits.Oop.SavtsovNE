```
using System;

public interface IFlyable
{
    void Fly();
}

public interface ISwimmable
{
    void Swim();
}

public interface IWalkable
{
    void Walk();
}

public interface IEatable
{
    void Eat();
}

abstract class Animal : IEatable
{
    public string Name { get; set; }

    public virtual void Eat()
    {
        Console.WriteLine($"{Name} ест...");
    }

    public abstract void MakeSound();
}

class Dog : Animal, IWalkable
{
    public override void MakeSound()
    {
        Console.WriteLine("Гав-гав!");
    }

    public void Walk()
    {
        Console.WriteLine($"{Name} гуляет.");
    }
}

class Cat : Animal, IWalkable
{
    public override void MakeSound()
    {
        Console.WriteLine("Мяу!");
    }

    public void Walk()
    {
        Console.WriteLine($"{Name} бродит.");
    }
}

class Bird : Animal, IFlyable, IWalkable
{
    public override void MakeSound()
    {
        Console.WriteLine("Чирик!");
    }

    public void Fly()
    {
        Console.WriteLine($"{Name} летает.");
    }

    public void Walk()
    {
        Console.WriteLine($"{Name} ходит.");
    }
}

class Fish : Animal, ISwimmable
{
    public override void MakeSound()
    {
        Console.WriteLine("Бульк!");
    }

    public void Swim()
    {
        Console.WriteLine($"{Name} плавает.");
    }
}

public class Food
{
    public string Type { get; set; }

    public Food(string type)
    {
        Type = type;
    }

    public void Serve()
    {
        Console.WriteLine($"Подана еда: {Type}");
    }
}

public class Heart
{
    public void Beat()
    {
        Console.WriteLine("Сердце бьётся...");
    }
}

public interface IAnimalBehavior
{
    void PerformAction();
}

public class DogBehavior : IAnimalBehavior
{
    private Dog _dog;

    public DogBehavior(Dog dog)
    {
        _dog = dog;
    }

    public void PerformAction()
    {
        _dog.MakeSound();
        _dog.Walk();
    }
}


public class AnimalManager
{
    private List<IAnimalBehavior> _behaviors = new List<IAnimalBehavior>();

    public void AddBehavior(IAnimalBehavior behavior)
    {
        _behaviors.Add(behavior);
    }

    public void ShowAllActions()
    {
        foreach (var behavior in _behaviors)
        {
            behavior.PerformAction();
        }
    }
}
```
