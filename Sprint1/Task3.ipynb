```

public class Animal
{
    private string _name;
    private int _age;
    private string _species;


    public static int TotalAnimals { get; private set; } = 0;


    public Animal()
    {
        TotalAnimals++;
    }


    public Animal(string name, int age, string species)
    {
        _name = name;
        _age = age;
        _species = species;
        TotalAnimals++;
    }


    public string Name
    {
        get { return _name; }
        set { _name = value; }
    }

    public int Age
    {
        get { return _age; }
        set
        {
            if (value >= 0)
                _age = value;
            else
                throw new ArgumentOutOfRangeException("Возраст не может быть отрицательным!");
        }
    }

    public string Species
    {
        get { return _species; }
        set { _species = value; }
    }


    public virtual void MakeSound()
    {
        Console.WriteLine($"{_name} издаёт звук.");
    }


    public void Sleep()
    {
        Console.WriteLine($"{_name} спит.");
    }


    public static void ShowTotalAnimals()
    {
        Console.WriteLine($"Всего животных: {TotalAnimals}");
    }
}


public class Dog : Animal
{
    private string _breed;


    public static int TotalDogs { get; private set; } = 0;

 
    public Dog(string name, int age, string breed) : base(name, age, "Собака")
    {
        _breed = breed;
        TotalDogs++;
    }


    public string Breed
    {
        get { return _breed; }
        set { _breed = value; }
    }


    public override void MakeSound()
    {
        Console.WriteLine($"{Name} лает: Гав-гав!");
    }


    public void PlayWith(Cat cat)
    {
        Console.WriteLine($"{Name} играет с {cat.Name}.");
    }
}

public class Cat : Animal
{
    private bool _isFluffy;


    public static int TotalCats { get; private set; } = 0;


    public Cat(string name, int age, bool isFluffy) : base(name, age, "Кошка")
    {
        _isFluffy = isFluffy;
        TotalCats++;
    }


    public bool IsFluffy
    {
        get { return _isFluffy; }
        set { _isFluffy = value; }
    }


    public override void MakeSound()
    {
        Console.WriteLine($"{Name} мурлычет: Мяу!");
    }

    public void Chase(Dog dog)
    {
        Console.WriteLine($"{Name} гонится за {dog.Name}.");
    }
}


class Program
{
    static void Main()
    {
        var dog = new Dog("Барбос", 3, "Овчарка");
        var cat = new Cat("Мурка", 2, true);

        dog.MakeSound();   
        cat.MakeSound();   

        dog.PlayWith(cat); 
        cat.Chase(dog);    

        Animal.ShowTotalAnimals(); 
        Dog.ShowTotalDogs();       
        Cat.ShowTotalCats();       
    }
}
```
